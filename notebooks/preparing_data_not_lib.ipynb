```python
import pandas as pd
import numpy as np

filename = "../data/HDFC_data.csv"
df = pd.read_csv(filename)
df.head()
```

```python
# columns = ["SERIES","PREVCLOSE","TOTALTRADES","LAST"]
# df.drop(columns=columns,inplace=True)
df = df.rename(columns = {"TOTTRDQTY":"Total Traded Quantity","TOTTRDVAL":"Total Traded Value","HIGH":"High","LOW":"Low","OPEN":"Open Price","CLOSE":"Close Price","SYMBOL":"Symbol","TIMESTAMP":"Date"})
df['Average Trading Price'] = df['Total Traded Value']/df['Total Traded Quantity']
df.head()
```

The data is now ready to be passed into the technical indicator script

```python
df.to_csv("../data/hdfc_usable.csv", index=False)
```
